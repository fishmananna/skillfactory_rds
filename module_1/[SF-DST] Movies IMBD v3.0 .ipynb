# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
import operator
result = {}
for index, row in data.iterrows():
    cinema_companies = row['production_companies'].split('|')
    for company in cinema_companies:
        if company in result:
            result[company]['count'] += 1
            result[company]['length'] += len(row.original_title)
        else:
            result[company] = {
                'count': 1,
                'length': len(row.original_title)
            }
for key, value in result.items():
    result[key] = value['length']/value['count']

sorted_r = sorted(result.items(), key=operator.itemgetter(1), reverse=True)
sorted_r
```

# 35. Какие актеры чаще всего снимаются в одном фильме вместе

1. Johnny Depp & Helena Bonham Carter
2. Hugh Jackman & Ian McKellen

4. Adam Sandler & Kevin James
5. Daniel Radcliffe & Rupert Grint

```python
import operator
result = {}
for index, row in data.iterrows():
    cinema_actors = row['cast'].split('|')
    for actor in cinema_actors:
        for partner in cinema_actors:
            if partner == actor:
                continue
            if actor not in result:
                result[actor] = {}
                
            result[actor][partner] = result[actor].get(partner, 0) + 1

for name in ['Johnny Depp', 'Helena Bonham Carter']:#проверяла что все верно и потом просто чекала по одному актеру из пары
    sorted_r = sorted(result[name].items(), key=operator.itemgetter(1), reverse=True)
    print(name, ': ', sorted_r, end='\n\n')
```

# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
import operator
data['profit'] = data.revenue - data.budget
result = {}
for index, row in data.iterrows():
    cinema_directors = row['director'].split('|')
    for director in cinema_directors:
        if director not in result:
            result[director] = {}
        if row.profit > 0:
            result[director]['prfilms'] = result[director].get('prfilms', 0) + 1
        result[director]['films'] = result[director].get('films', 0) + 1
        
for key, value in result.items():
    if 'prfilms' not in value:
        result[key] = 0
        continue
    result[key] = value['prfilms']/value['films']*100
    
    
sorted_r = sorted(result.items(), key=operator.itemgetter(1), reverse=True)
sorted_r  
```
